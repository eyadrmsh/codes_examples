## Code 

```python
# Load the tweets into a pandas DataFrame
tweets = df.copy()

chunksize = len(tweets)//4

# Use multiprocessing to perform sentiment analysis on all tweets
with mp.Pool(processes=4) as pool:
    tweets['clean_tweet'] = pool.map(clean_tweet, tweets['text'], chunksize)

# Use multiprocessing to perform sentiment analysis on all tweets
with mp.Pool(processes=4) as pool:
    tweets['sentiment_vader'] = pool.map(analyze_sentiment_vader, tweets['clean_tweet'], chunksize)

# Use multiprocessing to perform polarity analysis on all tweets
with mp.Pool(processes=4) as pool:
    tweets['emotions'] = pool.map(analyze_sentiment, tweets['clean_tweet'], chunksize)

tweets[['Happy','Angry', 'Surprise','Sad','Fear']] = tweets['emotions'].apply(lambda x: pd.Series(x))
```
